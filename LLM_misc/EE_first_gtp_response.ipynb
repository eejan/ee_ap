```python
from openai import OpenAI
```

```python
client = OpenAI(
  api_key=os.environ['OPENAI_API_KEY'],  # this is also the default, it can be omitted
)
```

```python
print(os.environ['OPENAI_API_KEY'])
```

### Text completions

```python
completion = client.completions.create(
    model='text-davinci-003', 
    prompt='Give me two reasons to learn OpenAI API with python?',
    max_tokens=300)
```

```python
aa = completion.choices[0]
```

```python
print(aa)
```

```python
print(aa.text)
```

### Chat completion 

```python
instruction='you are a chatbot'
content = 'Give me two reasons to learn OpenAI API with python?'

chat_response = client.chat.completions.create(model='gpt-3.5-turbo', messages=[{'role': 'system', 'content': instruction},
                                                                     {'role': 'user', 'content': content}]
                                        , temperature=0)
print(chat_response)
```

### word embedding 

```python
embedding_response = client.embeddings.create(input=[content], model='text-embedding-ada-002')
```

```python
embedding_response.data[0].embedding
```

```python
def get_embedding(text, model="text-embedding-ada-002"):
   text = text.replace("\n", " ")
   return client.embeddings.create(input = [text], model=model).data[0].embedding
```

```python
get_embedding(content)
```
